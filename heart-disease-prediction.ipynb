# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing dataset

```python
dataset = pd.read_csv('/kaggle/input/heart-failure-prediction/heart.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

# Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le6 = LabelEncoder()
le8 = LabelEncoder()
le10 = LabelEncoder()
X[ : , 1] = le1.fit_transform(X[ : , 1])
X[ : , 2] = le2.fit_transform(X[ : , 2])
X[ : , 6] = le6.fit_transform(X[ : , 6])
X[ : , 8] = le8.fit_transform(X[ : , 8])
X[ : , 10] = le10.fit_transform(X[ : , 10])
```

```python
print(X)
```

# Splitting dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(y_train)
```

```python
print(y_test)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

# Training Dataset

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression()
model_logistic.fit(X_train, y_train)
y_predict_log = model_logistic.predict(X_test)
```

```python
#SVM
from sklearn.svm import SVC
model_svm = SVC(probability=True)
model_svm.fit(X_train, y_train)
y_predict_svm = model_svm.predict(X_test)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors = 8)
model_knn.fit(X_train, y_train)
y_predict_knn = model_knn.predict(X_test)
```

```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(X_train, y_train)
y_predict_tree = model_tree.predict(X_test)
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier
model_random = RandomForestClassifier(n_estimators = 175)
model_random.fit(X_train, y_train)
y_predict_random = model_random.predict(X_test)
```

# Testing accuracy on different ML algorithms

```python
from sklearn.metrics import accuracy_score
logistic_accuracy = accuracy_score(y_test, y_predict_log)
SVM_accuracy = accuracy_score(y_test, y_predict_svm)
KNN_accuracy = accuracy_score(y_test, y_predict_knn)
tree_accuracy = accuracy_score(y_test, y_predict_tree)
random_accuracy = accuracy_score(y_test, y_predict_random)
```

```python
print(logistic_accuracy)
print(SVM_accuracy)
print(KNN_accuracy)
print(tree_accuracy)
print(random_accuracy)
```

# Plotting Accuracy of different ML algorithms

```python
plt.figure(figsize = (12, 6))
plt.bar("Logistic Regression", logistic_accuracy, width = 0.5)
plt.bar("Support Vector Machine", SVM_accuracy, width = 0.5)
plt.bar("KNeighbors", KNN_accuracy, width = 0.5)
plt.bar("Decision Tree", tree_accuracy, width = 0.5)
plt.bar("Random Forest", random_accuracy, width = 0.5)
plt.xlabel("Machine Learning Algorithm")
plt.ylabel('Accuracy')
```

# Confusion matrix and ROC curve

```python
#Logistic Regression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, roc_auc_score

# --- Confusion Matrix ---
cm = confusion_matrix(y_test, y_predict_log)
disp_labels = ['Class 0', 'Class 1']

# --- ROC values ---
y_proba = model_logistic.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
roc_auc = roc_auc_score(y_test, y_proba)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=disp_labels)
disp.plot(cmap='Blues', ax=ax[0], colorbar=False)
ax[0].set_title("Confusion Matrix")

# Plot ROC curve
ax[1].plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
ax[1].plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
ax[1].set_xlabel('False Positive Rate')
ax[1].set_ylabel('True Positive Rate')
ax[1].set_title('ROC Curve')
ax[1].legend(loc="lower right")
ax[1].grid()

plt.tight_layout()
plt.show()
```
